```python
#count of people opted into receiving marketing emails (0-not_opted,1-opted) 
data2['opted_in_to_mailing_list'].value_counts().rename_axis('Opted_or_not').reset_index(name='user_count')
```

```python
#The organization containing more number users are listed below with their organization_id and number of users
data2.org_id.value_counts().head(5).rename_axis('Org_id').reset_index(name='user_count')
```

```python
#Number of users created by which source categorized as follws:
import seaborn as sns
plt.figure(figsize=(10,5))
sns.countplot(data2['creation_source']);
plt.xticks(rotation=90);
```

```python
#The number of users who are in regular marketing email drip are(1 - on marketing email drip, 0 - not on marketing email drip ) 
sns.countplot(data2['enabled_for_marketing_drip']);
plt.ylabel('No.of users')
plt.xticks(rotation=90);
```

```python
#The users who invited more people to join are listed as follows with their user_id and number of people joined through them :
data2['invited_by_user_id'].value_counts().head(5).rename_axis('User_id').reset_index(name='Number_of_users')
```
